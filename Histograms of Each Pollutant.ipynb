# Histograms of Each Pollutant

```python
import pandas as pd
import numpy as np
import seaborn as sns
import re
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pollutants_data = pd.read_csv('BA_MM_public_use_data.csv')
```

# NO

```python
sns.histplot(data = pollutants_data, x = 'NO')
```

# NO2

```python
sns.histplot(data = pollutants_data, x = 'NO2')
```

# BC

```python
sns.histplot(data = pollutants_data, x = 'BC')
```

# UFP

```python
sns.histplot(data = pollutants_data, x = 'UFP')
```
